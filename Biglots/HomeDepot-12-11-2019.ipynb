```python
import requests

from fake_useragent import UserAgent
from bs4 import BeautifulSoup
```

```python
ua = UserAgent()
header = {'User-Agent': str(ua.random)}
result = requests.get('https://www.homedepot.com/s/Badminton%2520-%2520Net%2520Sportcraft?NCNI-5', headers=header).content
```

```python
soup = BeautifulSoup(result, 'html.parser')
```

```python
base_url = 'https://www.homedepot.com'
```

```python
print(soup)
```

```python
product_div = soup.find_all('div', class_='plp-pod--default')
```

```python
product_div
```

```python
 for single_product_div in product_div:
        image_div = single_product_div.find('a', attrs={'data-podaction': 'product image'})
        anchor_tag = image_div
```
